```python
# Write a program using a while loop to print all digits of a number one by one

n = int(input("Enter number : "))
while n > 0 : 
    digit = n % 10     # n % 10 last number ko nikalta hai
    n //= 10           # n //= 10 last number ko remove kr deta hai
    print(digit)
```

```python
# Write a program to find the sum of digits of a number using a while loop

n = int(input("Enter number : "))
sum = 0
while n > 0 :
    digit = n % 10
    sum += digit
    n //= 10
    print("The sum of digits : ",sum)
```

```python
# Write a program to count how many digits are in a given number using a while loop

n = int(input("Enter a number : "))
count = 0
while n > 0 :
    count += 1
    n //= 10
print(count)
```

```python
# Write a program to reverse a number using a while loop

n = int(input("enter a number : "))
reverse = 0
while n > 0 :
    reverse = reverse * 10 + n % 10       # yeh reverse ke liye use hota hai 
    n //= 10
    print("The reverse number : ", reverse)
```

```python
# Write a program to check whether a number is a palindrome using a while loop (121 --> 121)

n = int(input("Enter number : "))
temp = n
reverse = 0

while n > 0 :
    reverse = reverse * 10 + n % 10
    n //= 10

if temp == reverse :
    print("Number is palindrome")
else :
    print("Number is not palindrome")
```

```python
# Write a program to find the product of digits of a number using a while loop

n = int(input("Enter number : "))
product = 1
while n > 0 :
    product = product * (n % 10)
    n //= 10
    print("The product of digits : ", product)
```

```python
# Write a program to find the largest digit in a number using a while loop

n = int(input("Enter number : "))
largest = 0
while n > 0 :
    digit = n % 10
    if(digit > largest):
        largest = digit
    n //= 10
print("The largest number : ", largest)
```

```python
# Write a program to find the smallest digit in a number using a while loop

n = int(input("Enter number : "))
smallest = 9
while n > 0 :
    digit = n % 10
    if(digit < smallest) :
        smallest = digit
    n //= 10
print("The smallest number : ", smallest)
```

```python
# Write a program to check whether a number is an Armstrong number using a while loop

n = int(input("Enter number : "))
temp = n
sum = 0
while n > 0 :
    digit = n % 10
    sum += digit ** 3
    n //= 10 
if(temp == sum) :
    print("Number is armstrong")
else :
    print("Number is not armstrong")
```

```python
# Write a program to remove all zeros from a number using a while loop

n = int(input("Enter number : "))
new = 0
place = 1
while n > 0 : 
    digit = n % 10
    if digit != 0 :
        new += digit * place
        place *= 10
    n //= 10
print(new)
```

```python
# Write a program to count how many even and odd digits are present in a number using a while loop

n = int(input("Enter number : "))
even = 0
odd = 0
while n > 0 :
    digit = n % 10
    if digit % 2 == 0 :
        even += 1
    else : 
        odd += 1
    n //= 10
print("Count of even numbers : ", even)
print("Count of odd numbers : ", odd)
```

```python
# Write a program to check whether a number contains a specific digit (for example, 5) using a while loop

n = int(input("Enter number : "))
found = False
while n > 0 :
    if n % 10 == 5 :
        found = True
        break
    n //= 10
if found == True :
    print("Number is found")
else :
    print("Number is not found")
```

```python
# Write a program to calculate the sum of even digits and sum of odd digits separately using a while loop

n = int(input("Enter number : "))
even_sum = 0
odd_sum = 0
while n > 0 :
    digit = n % 10
    if n % 2 == 0 :
        even_sum += digit
    else :
        odd_sum += digit
    n //= 10
print("The sum of even number : ", even_sum)
print("The sum of odd number : ", odd_sum)
```

```python
# Write a program to create a new number by squaring each digit of a given number using a while loop

n = int(input("Enter number : "))
result = ""
while n > 0 :
    digit = n % 10
    result = str(digit ** 2) + result
    n //= 10
print(result)
```

```python
# Write a program to check whether a number is a perfect number using a while loop

n = int(input("Enter number : "))
i = 1
sum = 0
while i < n :
    if n % i == 0 :
        sum += i
    i += 1
if sum == n :
    print("Perfect number")
else :
    print("Not a perfect number")
```

```python
# Write a program to print the Fibonacci series up to N terms using a while loop

n = int(input("Enter terms : "))
a = 0
b = 1
i = 1
while i <= n :
    print(a)
    a , b = b , a + b
    i += 1
```

```python
# Write a program to find the factorial of a number using a while loop

n = int(input("Enter number : "))
factorial = 1
while n > 0 :
    factorial *= n
    n -= 1
print("Factorial of number : ", factorial)
```

```python
# Write a program to calculate the power of a number (a^b) using a while loop

a = int(input("Enter base : "))
b = int(input("Enter power : "))
result = 1
while b > 0 :
    result *= a
    b -= 1
print(result)
```

```python
# Write a program to find the GCD of two numbers using a while loop

a = int(input("Enter a : "))
b = int(input("Enter b : "))
while a != b :
    if a > b :
        a -= b
    else :
        b -= a
print("The HCF / GCD is : ", a)
```
