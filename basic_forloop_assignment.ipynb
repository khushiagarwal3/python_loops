```python
# for loop basic assignment 
```

```python
# Print numbers from 1 to 10

for i in range(1,11) :
    print(i)
```

```python
# Print numbers from 10 to 1

for i in range(10,0,-1) :
    print(i)
```

```python
# Print all even numbers from 1 to 50

for i in range(2, 51, 2) :
    print(i)
```

```python
# Print all odd numbers from 1 to 50

for i in range(1, 51, 2) :
    print(i)
```

```python
# Print numbers from 0 to 100 with a step of 10

for i in range(10, 101, 10) :
    print(i)
```

```python
# Print numbers from 5 to 50 with a step of 5

for i in range(5, 51, 5) :
    print(i)
```

```python
# Print numbers from 100 to 0 with a step of -5

for i in range(100, 0, -5) :
    print(i)
```

```python
# Print the square of numbers from 1 to 10

for i in range(1, 11) :
    print(i * i)
```

```python
# Print the cube of numbers from 1 to 5

for i in range(1, 6) :
    print(i * i * i)
```

```python
# Find the sum of numbers from 1 to 100

total = 0
for i in range(1, 101) :
    total += i
print("The sum of number : ",total)
```

```python
# Find the sum of even numbers from 1 to 50

total = 0
for i in range(2, 51, 2) :
    total += i
print("The sum of even number : ", total)
```

```python
# Print the table of 7

for i in range(1,11) :
    print(7 * i)
```

```python
# Print numbers between 1 and 100 that are divisible by 4

for i in range(1, 101) :
    if(i % 4 == 0) :
        print(i)
```

```python
# Print numbers from 1 to 30 but skip multiples of 3

for i in range(1, 31) :
    if(i % 3 == 0) :
        continue
    print(i)
```

```python
# Print numbers from 1 to 50 and stop when the number becomes 25

for i in range(1, 51) :
    if(i == 25) :
        break
    print(i)
```

```python
# Print numbers from 20 to 1

for i in range(20, 0, -1) :
    print(i)
```

```python
# Print all multiples of 6 between 1 and 100

for i in range(1, 101) :
    if(i % 6 == 0) :
        print(i)
```

```python
# Print numbers from 2 to 20 with a step of 2

for i in range(2, 21, 2) :
    print(i)
```

```python
# Print numbers from 1 to 100 in reverse order

for i in range(100, 0, -1) :
    print(i)
```

```python
# Count how many numbers are there from 1 to 100

count = 0
for i in range(1, 101) :
    count += 1
print(count)
```
